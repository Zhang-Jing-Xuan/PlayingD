```python
import tarfile
def untar(fname, dirs):
    """
    解压tar.gz文件
    :param fname: 压缩文件名
    :param dirs: 解压后的存放路径
    :return: bool
    """
    try:
        t = tarfile.open(fname)
        t.extractall(path = dirs)
        return True
    except Exception as e:
        print(e)
        return False
    
# source="cifar-10-python.tar.gz"
# target="./data"
source="flower_photos.tgz"
target="./flower_data"
untar(source,target)
```
