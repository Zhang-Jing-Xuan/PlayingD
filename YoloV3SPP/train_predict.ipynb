```python
import tarfile
import os
source="/home/mist/VOCtrainval_11-May-2012.tar"
target="/home/mist"
def un_tar(file_name):
    tar = tarfile.open(file_name)
    names = tar.getnames()
    if os.path.isdir(target):
        pass
    else:
        os.mkdir(target)
    #因为解压后是很多文件，预先建立同名目录
    for name in names:
        tar.extract(name, target)
    tar.close()
    print("解压完成")
un_tar(source)
```
